```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
x_train=np.array([1.0,2.0])
y_train=np.array([300.0,500.0])
print('x_train = ',x_train)
print('y_train = ',y_train)
```

```python
print('x_train.shape : ',x_train.shape )
m=x_train.shape[0]
print('The number of training examples are ',m)
```

```python
# m can also be calculated as
m=len(x_train)
print('The number of training examples are ',m)
```

```python
i = 0 # Change this to 1 to see (x^1, y^1)

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train,y_train,marker='x',c='r')
plt.title('Housing Prices')
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
np.zeros(5)
```

```python
def compute_model_output(x,w,b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=w*x[i]+b
        
    return f_wb
compute_model_output(x_train,w,b)
```

```python
tmp_f_wb=compute_model_output(x_train,w,b)
# to create a line plot(model prediction)
plt.plot(x_train,tmp_f_wb,c='b',label='Our Prediction')
# to create a scatter plot (to plot the data points)
plt.scatter(x_train,y_train,marker='x',c='r',label='Actual Values')
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
tmp_f_wb=compute_model_output(x_train,200,100)
# to create a line plot(model prediction)
plt.plot(x_train,tmp_f_wb,c='b',label='Our Prediction')
# to create a scatter plot (to plot the data points)
plt.scatter(x_train,y_train,marker='x',c='r',label='Actual Values')
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
# Now these values of w and b fits our data well
```
